# Assignment 11
Show that the relation R consisting of all pairs (x, y) such
that x and y are bit strings of length three or more that
agree in their first three bits is an equivalence relation on
the set of all bit strings of length three or more.

# Answer

## Definition of Relation $ R $
The relation $ R $ is defined on the set of all bit strings of length three or more. For two bit strings $ x $ and $ y $, $ (x, y) \in R $ if and only if $ x $ and $ y $ agree in their first three bits.

To show that $ R $ is an equivalence relation, we need to verify the following properties:
1. Reflexivity
2. Symmetry
3. Transitivity

---

## 1. Reflexivity
A relation $ R $ is reflexive if $ (x, x) \in R $ for all $ x $ in the set.

For any bit string $ x $, the first three bits of $ x $ clearly agree with themselves. Thus, $ (x, x) \in R $.

**Conclusion**: $ R $ is reflexive.

---

## 2. Symmetry
A relation $ R $ is symmetric if $ (x, y) \in R $ implies $ (y, x) \in R $.

If $ (x, y) \in R $, then the first three bits of $ x $ agree with the first three bits of $ y $. Since equality is symmetric, the first three bits of $ y $ also agree with the first three bits of $ x $. Thus, $ (y, x) \in R $.

**Conclusion**: $ R $ is symmetric.

---

## 3. Transitivity
A relation $ R $ is transitive if $ (x, y) \in R $ and $ (y, z) \in R $ imply $ (x, z) \in R $.

If $ (x, y) \in R $, then $ x $ and $ y $ agree on their first three bits. Similarly, if $ (y, z) \in R $, then $ y $ and $ z $ agree on their first three bits. Since the first three bits of $ y $ agree with both $ x $ and $ z $, it follows that $ x $ and $ z $ also agree on their first three bits. Thus, $ (x, z) \in R $.

**Conclusion**: $ R $ is transitive.

---

## Final Conclusion
Since $ R $ satisfies reflexivity, symmetry, and transitivity, it is an equivalence relation on the set of all bit strings of length three or more.


# Assignment 15
Let R be the relation on the set of ordered pairs
of positive integers such that $((a, b), (c, d)) \in R$ if and
only if a + d = b + c. Show that R is an equivalence
relation.

### Answer

#### Definition of Relation $ R $
The relation $ R $ is defined on the set of ordered pairs of positive integers. For two pairs $(a, b)$ and $(c, d)$, 
$$
((a, b), (c, d)) \in R \iff a + d = b + c.
$$

To show that $ R $ is an equivalence relation, we must verify the following properties:
1. Reflexivity
2. Symmetry
3. Transitivity

---

#### 1. Reflexivity
A relation $ R $ is reflexive if $((a, b), (a, b)) \in R$ for all $(a, b)$ in the set.

For any $(a, b)$, we have:
$$
a + b = b + a.
$$
Thus, $((a, b), (a, b)) \in R$.

**Conclusion**: $ R $ is reflexive.

---

#### 2. Symmetry
A relation $ R $ is symmetric if $((a, b), (c, d)) \in R$ implies $((c, d), (a, b)) \in R$.

If $((a, b), (c, d)) \in R$, then:
$$
a + d = b + c.
$$
By the symmetry of equality, we also have:
$$
c + b = d + a.
$$
Thus, $((c, d), (a, b)) \in R$.

**Conclusion**: $ R $ is symmetric.

---

#### 3. Transitivity
A relation $ R $ is transitive if $((a, b), (c, d)) \in R$ and $((c, d), (e, f)) \in R$ imply $((a, b), (e, f)) \in R$.

If $((a, b), (c, d)) \in R$, then:
$$
a + d = b + c.
$$
If $((c, d), (e, f)) \in R$, then:
$$
c + f = d + e.
$$
Adding these equations gives:
$$
(a + d) + (c + f) = (b + c) + (d + e).
$$
Simplifying, we find:
$$
a + f = b + e.
$$
Thus, $((a, b), (e, f)) \in R$.

**Conclusion**: $ R $ is transitive.

---

#### Final Conclusion
Since $ R $ satisfies reflexivity, symmetry, and transitivity, it is an equivalence relation on the set of ordered pairs of positive integers.

\subsubsection*{Exercise 10 (9.5):}
Suppose that $A$ is a nonempty set and $R$ is an equivalence
relation on $A$. Show that there is a function $f$ with $A$ as its
domain such that $(x, y) \in R$ if and only if $f(x) = f(y)$.

---

### Proof:

1. **Properties of Equivalence Relation:**
   Since $R$ is an equivalence relation on $A$, it satisfies the following properties:
   - Reflexivity: $(x, x) \in R$ for all $x \in A$.
   - Symmetry: If $(x, y) \in R$, then $(y, x) \in R$.
   - Transitivity: If $(x, y) \in R$ and $(y, z) \in R$, then $(x, z) \in R$.

2. **Partition Induced by $R$:**
   The equivalence relation $R$ partitions the set $A$ into disjoint equivalence classes. For each $x \in A$, let $[x]_R$ denote the equivalence class of $x$ under $R$, i.e.:
   $$
   [x]_R = \{y \in A \mid (x, y) \in R\}.
   $$

   These equivalence classes are disjoint, and every element of $A$ belongs to exactly one equivalence class.

3. **Defining the Function $f$:**
   Define a function $f: A \to A$ such that:
   $$
   f(x) = [x]_R,
   $$
   where $f(x)$ maps each element $x \in A$ to its equivalence class $[x]_R$.

4. **Verify the Condition:**
   To show that $(x, y) \in R$ if and only if $f(x) = f(y)$:
   - If $(x, y) \in R$, then $x$ *and $y$ belong to the same equivalence class, so $[x]_R = [y]_R$. Hence, $f(x) = f(y)$.
   - Conversely, if $f(x) = f(y)$, then $x$ and $y$ belong to the same equivalence class, which means $(x, y) \in R$.

5. **Conclusion:**
   The function $f(x) = [x]_R$ satisfies the condition that $(x, y) \in R$ if and only if $f(x) = f(y)$.

Thus, the function $f$ exists and has the desired properties.
