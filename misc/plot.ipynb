```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # This import registers the 3D projection

# Define the grid for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute the function f(x,y)
Z = np.sqrt(X**2 + Y**2)

# Create a figure and a 3D Axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title('Plot of $f(x,y)=\\sqrt{x^2+y^2}$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')

# Add a color bar which maps values to colors
fig.colorbar(surface, shrink=0.5, aspect=5)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constant level c (for example, c = 3)
c = 3

# Define the grid for x and y
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

# Compute the function f(x,y)
Z = np.sqrt(X**2 + Y**2)

# Create a plot
plt.figure(figsize=(6,6))

# Plot the level curve where f(x,y) = c using contour
contour = plt.contour(X, Y, Z, levels=[c], colors='red', linewidths=2)

# Optionally add labels to the contour line
plt.clabel(contour, inline=True, fontsize=10, fmt={c: f'f(x,y)={c}'})

# Set plot labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Level Curve for $f(x,y)=\\sqrt{{x^2+y^2}}$ at level $c={c}$')
plt.grid(True)
plt.axis('equal')  # Ensure the aspect ratio is equal to correctly display the circle

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Registers the 3D projection

# Define the domain for x and y
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)

# Define the function f(x, y) = x^2 + y^2
Z = X**2 + Y**2

# Create a figure with two subplots (side by side)
fig = plt.figure(figsize=(14, 6))

# --------------------------
# Left subplot: 3D Surface Plot
# --------------------------
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
surface = ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax1.set_title(r'3D Surface Plot of $f(x,y)=x^2+y^2$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x,y)')
fig.colorbar(surface, ax=ax1, shrink=0.5, aspect=10)

# --------------------------
# Right subplot: 2D Contour Plot
# --------------------------
ax2 = fig.add_subplot(1, 2, 2)
# Plot filled contours (scalar field)
contour = ax2.contourf(X, Y, Z, levels=50, cmap='viridis')
ax2.set_title(r'2D Contour Plot (Scalar Field) of $f(x,y)=x^2+y^2$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True)
ax2.set_aspect('equal', 'box')  # Ensure the aspect ratio is equal
fig.colorbar(contour, ax=ax2)

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter range for plotting lines
t = np.linspace(-10, 10, 300)

# --- Kernel: ker(A) = span{(2, 1)} ---
kernel_vector = np.array([2, 1])
kernel_line = np.outer(kernel_vector, t)  # Each column: [2*t, 1*t]

# --- Column space: col(A) = span{(1, -1)} ---
col_vector = np.array([1, -1])
s = np.linspace(-10, 10, 300)
col_line = np.outer(col_vector, s)

# --- Level Curve ---
# Choose a level c, e.g., c = 1.
c = 1
# A particular solution: x0 = (c, 0) gives f(x0) = (c, -c)
x0 = np.array([c, 0])
# The level curve is x0 + ker(A)
level_line = x0.reshape(2, 1) + np.outer(kernel_vector, t)

# --- Plotting ---
plt.figure(figsize=(8, 8))
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)

# Plot kernel (red solid line)
plt.plot(kernel_line[0, :], kernel_line[1, :], 'r-', lw=2,
         label=r'$\ker(A)=\operatorname{span}\{(2,1)\}$')

# Plot column space (blue dashed line)
plt.plot(col_line[0, :], col_line[1, :], 'b--', lw=2,
         label=r'$\operatorname{col}(A)=\operatorname{span}\{(1,-1)\}$')

# Plot level curve (green solid line)
plt.plot(level_line[0, :], level_line[1, :], 'g-', lw=3,
         label=r'Level curve: $x_0 + t\,(2,1)$')

# Mark the particular point x0 used for the level curve
plt.plot(x0[0], x0[1], 'ko', markersize=8,
         label=r'$x_0=(c,0)$')

# Annotate the plot
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title('Visualization of $f(x)=Ax$: Kernel, Column Space, and a Level Curve')
plt.legend(loc='upper right')
plt.grid(True)
plt.axis('equal')  # Use equal scaling on both axes

plt.show()
```

```python
def er_primtal(x: int):
    """Returnerer True, hvis num er et primtal, ellers False."""
    if x < 2:
        return False
    # Tjekker divisibilitet fra 2 op til kvadratroden af num
    for i in range(2, int(x ** 0.5) + 1):
        if x % i == 0:
            return False
    return True

def f(n: int):
    """
    Returnerer det næste primtal større end n.
    """
    kandidat = n + 1
    while True:
        if er_primtal(kandidat):
            return kandidat
        kandidat += 1

# Test af funktionen:
print("f(3) =", f(3))    # Forventet output: 5
print("f(10) =", f(10))  # Forventet output: 11
```

```python
def er_primtal(x: int) -> bool:
    """Returnerer True, hvis x er et primtal, ellers False."""
    if x < 2:
        return False
    # Tjekker divisibilitet fra 2 op til kvadratroden af x
    for i in range(2, int(x ** 0.5) + 1):
        if x % i == 0:
            return False
    return True

def f(n: float) -> int:
    """
    Returnerer det næste primtal, der er strengt større end n.

    Parameter:
      n (float): Et reelt tal. Funktionen returnerer det mindste primtal p, hvor p > n.

    Bemærk: Hvis n er negativt, returneres det mindste primtal (2).
    """
    # Start fra det næste heltal over n. Hvis n er et heltal, sikrer vi os, at vi starter med n+1.
    kandidat = max(2, int(n) + 1)
    if kandidat <= n:
        kandidat += 1

    # Iterér indtil vi finder et primtal.
    while True:
        if er_primtal(kandidat):
            return kandidat
        kandidat += 1

# Test af funktionen:
print("f(3) =", f(3))       # Forventet output: 5
print("f(10) =", f(10))     # Forventet output: 11
print("f(10.2) =", f(10.2)) # Forventet output: 11
print("f(-5) =", f(-5))     # Forventet output: 2
```
