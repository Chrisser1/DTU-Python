```python
x1,x2 = symbols('x1 x2')
x = Matrix([x1,x2])

A1 = Matrix([[11,-12],[-12,4]])
A2 = Matrix([[11,0],[-24,4]])
A3 = Matrix([[73/5,-36/5],[-36/5,52/5]])

b1 = Matrix([-20,40])
b2 = b1
b3 = Matrix([-44,8])

c = Matrix([-60])

x,A1,A2,A3,b1,b2,b3,c
```

```python
q1 = x.T*A1*x + b1.T*x + c
simplify(q1)
```

```python
q2 = x.T*A2*x + b2.T*x + c
simplify(q2)
```

```python
q3 = x.T*A3*x + b3.T*x + c
simplify(q3)
```

```python
q = Matrix([q1,q2,q3])
simplify(q)
```

```python
# We check if q1 and q2 are equivalent
simplify(q1-q2)
```

```python
from dtumathtools.dtuplot import plot3d, plot_contour
```

```python
# plot q1 and q3
plot3d(q1, (x1, -10, 10), (x2, -10, 10), title='q1', xlabel='x1', ylabel='x2')
plot3d(q3, (x1, -10, 10), (x2, -10, 10), title='q3', xlabel='x1', ylabel='x2')
```

```python
plot_contour(q1, (x1, -10, 10), (x2, -10, 10), is_filled=False, title='q1', xlabel='x1', ylabel='x2')
plot_contour(q3, (x1, -5, 5), (x2, -5, 5), is_filled=False, title='q1', xlabel='x1', ylabel='x2')
```

```python
from sympy import symbols, diff
from dtumathtools.dtuplot import plot_vector, plot_contour

# Define symbolic variables
x, y = symbols('x y')

# Define the function f(x, y)
f = x**2 - 2*y

# Compute the gradient of f
grad_f = [diff(f, x), diff(f, y)]

# Plot the contour levels (level curves) of f
p1 = plot_contour(f, (x, -3, 3), (y, -3, 3), show=False)

# Plot the gradient vector field over the contours
p2 = plot_vector(grad_f, (x, -3, 3), (y, -3, 3), use_cm=False, show=False)

# Display the combined plot
(p1 + p2).show()
```
